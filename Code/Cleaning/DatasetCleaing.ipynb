### Changing the names of the columns

Attribute Information:
	
1. loc             : numeric % McCabe's line count of code

2. v(g)            : numeric % McCabe "cyclomatic complexity"

3. ev(g)           : numeric % McCabe "essential complexity"

4. iv(g)           : numeric % McCabe "design complexity"

5. n               : numeric % Halstead total operators + operands

6. v               : numeric % Halstead "volume"

7. l               : numeric % Halstead "program length"

8. d               : numeric % Halstead "difficulty"

9. i               : numeric % Halstead "intelligence"

10. e               : numeric % Halstead "effort"

11. b               : numeric % Halstead "bugs"

12. t               : numeric % Halstead's time estimator

13. lOCode          : numeric % Halstead's line count

14. lOComment       : numeric % Halstead's count of lines of comments

15. lOBlank         : numeric % Halstead's count of blank lines

16. lOCodeAndComment: numeric % Halstead's count of lines of code and comments

17. uniq_Op         : numeric % unique operators

18. uniq_Opnd       : numeric % unique operands

19. total_Op        : numeric % total operators

20. total_Opnd      : numeric % total operands

21: branchCount     : numeric % of the flow graph

22. defects         : {false,true} % module has/has not one or more reported defects

```python
df = df.rename(columns={
    'loc':'McCabe "line count of code"',
    'v(g)':'McCabe "cyclomatic complexity"',
    'ev(g)':'McCabe "essential Complexity"',
    'iv(g)':'McCabe "design complexity"',
    'n':'Halstead "total operators + operands"',
    'v':'Halstead "volume"',
    'l':'Halstead "program length"',
    'd':'Halstead "difficulty"',
    'i':'Halstead "intelligence"',
    'e':'Halstead "effort"',
    'b':'Halstead "bugs"',
    't':'Halstead "time estimator"',
    'lOCode':'Halstead "line Count"',
    'lOComment':'Halstead "count of lines of comments"',
    'lOBlank':'Halstead "count of blank lines"',
    'lOCodeAndComment': 'Halstead "count of lines of code and comments"',
    'uniq_Op':'Unique operators',
    'uniq_Opnd':'Unique operands',
    'total_Op':'Total operators',
    'total_Opnd':'Total operands',
    'branchCount':'Count of all branches',
    'defects':'Defects'
})
```

New names have been assigned to the columns

```python
df.info()
```

In the dataset above, some columns that contain numerical values are incorrectly classified as 'objects'. Therefore, we need to convert these data types to numeric values.

### Data Conversion

```python
for columns in df.columns:
    if df[columns].dtype == 'object':
        df[columns] = pd.to_numeric(df[columns], errors='coerce') 
```

```python
df.info()
```

We observe that some columns contain NaN values. Therefore, we need to identify where these NaN values are located.

```python
nan_rows = df[df.isna().any(axis=1)]
nan_rows
```

Given the situation, we have decided to remove the rows containing NaN values.

### Data Cleaning

```python
df = df.dropna()
df.info()
```
